```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("trainn.csv")
```

```python
import numpy as np
```

```python
data= data.dropna()
```

```python
data.shape
```

```python
data.head(10)
```

```python
data.reset_index(inplace=True)
```

```python
data.head(10)
```

```python
data=data.iloc[:,2:]
```

```python
data
```

```python
#lowering the text
data["title"]=data["title"].apply(lambda x: x.lower())
```

```python
#cleaning the text
data["title"]=data.title.str.replace('[^a-zA-Z ]', '')
```

```python
df=data.iloc[:,[0,3]]
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
train_df,eval_df = train_test_split(df,test_size = 0.05)
```

```python
pip install simpletransformers
```

```python

from simpletransformers.classification import ClassificationModel


# Create a TransformerModel
model = ClassificationModel('electra', 'google/electra-small-discriminator', num_labels=2, 
                            args={'reprocess_input_data': True, 'overwrite_output_dir': True},use_cuda=False)
model.args.num_train_epochs=6
```

```python
model.train_model(train_df)
```

```python
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```

```python
print(result)
print(model_outputs)
```

```python
lst = []
for arr in model_outputs:
    lst.append(np.argmax(arr))
```

```python
true = eval_df['label'].tolist()
predicted = lst
```

```python
import sklearn
mat = sklearn.metrics.confusion_matrix(true , predicted)
mat
```

```python
print(sklearn.metrics.classification_report(true,predicted,target_names=['real','fake']))
```
